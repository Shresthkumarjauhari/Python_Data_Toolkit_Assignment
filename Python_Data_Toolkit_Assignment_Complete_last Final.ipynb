# Data Toolkit Assignment

```python
# Q1. What is NumPy, and why is it widely used in Python?
# NumPy provides fast numerical arrays and mathematical operations. Example: np.array([1,2,3]).
```

```python
# Q2. How does broadcasting work in NumPy?
# Broadcasting performs operations on arrays of different shapes automatically. Example: arr+5.
```

```python
# Q3. What is a Pandas DataFrame?
# A DataFrame is a 2D labeled table in Pandas. Example: pd.DataFrame({'A':[1,2]}).
```

```python
# Q4. Explain the use of the groupby() method in Pandas.
# groupby() groups data for aggregation. Example: df.groupby('Dept').sum().
```

```python
# Q5. Why is Seaborn preferred for statistical visualizations?
# Seaborn creates attractive statistical graphs with less code. Example: sns.histplot().
```

```python
# Q6. What are the differences between NumPy arrays and Python lists?
# NumPy arrays are faster and homogeneous; lists are flexible. Example: np.array vs list.
```

```python
# Q7. What is a heatmap, and when should it be used?
# A heatmap shows values using colors. Example: correlation matrix.
```

```python
# Q8. What does the term 'vectorized operation' mean in NumPy?
# Vectorized operations process entire arrays without loops. Example: arr*2.
```

```python
# Q9. How does Matplotlib differ from Plotly?
# Matplotlib is static; Plotly is interactive. Example: zoomable Plotly charts.
```

```python
# Q10. What is the significance of hierarchical indexing in Pandas?
# Hierarchical indexing allows multiple index levels. Example: MultiIndex.
```

```python
# Q11. What is the role of Seaborn's pairplot() function?
# pairplot() visualizes relationships among variables. Example: sns.pairplot(df).
```

```python
# Q12. What is the purpose of the describe() function in Pandas?
# describe() gives summary statistics. Example: df.describe().
```

```python
# Q13. Why is handling missing data important in Pandas?
# It prevents incorrect analysis. Example: fillna().
```

```python
# Q14. What are the benefits of using Plotly for data visualization?
# Plotly provides interactive visualizations. Example: px.bar().
```

```python
# Q15. How does NumPy handle multidimensional arrays?
# NumPy supports N-dimensional arrays. Example: np.zeros((2,3,4)).
```

```python
# Q16. What is the role of Bokeh in data visualization?
# Bokeh creates interactive web visualizations. Example: line chart.
```

```python
# Q17. Explain the difference between apply() and map() in Pandas.
# apply() works on Series/DataFrame; map() mainly on Series. Example: s.map().
```

```python
# Q18. What are some advanced features of NumPy?
# Advanced features include broadcasting, vectorization, linear algebra, random module.
```

```python
# Q19. How does Pandas simplify time series analysis?
# Pandas provides datetime indexing and resampling. Example: pd.to_datetime().
```

```python
# Q20. What is the role of a pivot table in Pandas?
# Pivot tables summarize data. Example: pd.pivot_table().
```

```python
# Q21. Why is NumPy's array slicing faster than Python's list slicing?
# Array slicing returns views, making it faster than list copies.
```

```python
# Q22. What are some common use cases for Seaborn?
# Seaborn is used for heatmaps, pairplots, distributions, regression and categorical plots.
```
